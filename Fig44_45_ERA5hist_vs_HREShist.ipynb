```python
# -*- encoding: utf-8 -*- #

''' 对比 ERA5 AAM hist 和 HRES AAM hist 之间的差异
'''  # pylint: disable = C0103
from TimeStamp.series import Series, ModLS
from AAM_Compare.src.aam import (
    get_AAM_ERA3h_hist_2022,
    get_AAM_GFZ_hist
)
Per = [ 14.9589314, 15.0000000, 15.0410686,
        28.4397295, 28.9841042, 29.5284789,
        29.9589333, 30.0000000, 30.0410667,
        44.9589300, 45.0000000, 45.0410700]  # deg/hour
Per = [ p/15.0 for p in Per ]   # cpd
MT = ModLS.setup(periods=Per, poly_deg=None, pers_unit='day')

# 加载数据+预处理
HRESp, HRESm = get_AAM_GFZ_hist(year_range=2022)
HRESp.de_mean(inplace=True)
HRESm.de_mean(inplace=True)
# 扣除12个潮汐项
MT_HRESp, MT_HRESm = MT.solve([HRESp, HRESm])
# 滑动平均， 扣除后AAM低频信号符合得更好
HRESp_mv  = MT_HRESp.residual.dilute(times=1, method='mv3', width=9)
HRESm_mv  = MT_HRESm.residual.dilute(times=1, method='mv3', width=9)
HRESp_res = HRESp - HRESp_mv
HRESm_res = HRESm - HRESm_mv

# 加载数据+预处理
ERA5p, ERA5m = get_AAM_ERA3h_hist_2022()
ERA5p.de_mean(inplace=True)
ERA5m.de_mean(inplace=True)
# 扣除12个潮汐项
MT_ERA5p, MT_ERA5m = MT.solve([ERA5p, ERA5m])
# 滑动平均， 扣除后AAM低频信号符合得更好
ERA5p_mv = MT_ERA5p.residual.dilute(times=1, method='mv3', width=9)
ERA5m_mv = MT_ERA5m.residual.dilute(times=1, method='mv3', width=9)
ERA5p_res = ERA5p - ERA5p_mv
ERA5m_res = ERA5m - ERA5m_mv
```
